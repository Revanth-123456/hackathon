```python
import pandas as pd

# Load CSV file
orders_df = pd.read_csv("orders.csv")

# Check data
print(orders_df.head())
print(orders_df.info())
```

```python
# Load JSON file
users_df = pd.read_json("users.json")

# Check data
print(users_df.head())
print(users_df.info())
```

```python
import sqlite3

# Create database connection
conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()
```

```python
# Read SQL file
with open("restaurants.sql", "r") as file:
    sql_script = file.read()

# Execute SQL script
cursor.executescript(sql_script)

conn.commit()
```

```python
restaurants_df = pd.read_sql_query(
    "SELECT * FROM restaurants",
    conn
)

print(restaurants_df.head())
print(restaurants_df.info())
```

```python
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)

print(orders_users_df.head())
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)

print(final_df.head())
```

```python
final_df.to_csv(
    "final_food_delivery_dataset.csv",
    index=False
)

print("Final dataset saved successfully!")
```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")
```

```python
gold_city_revenue = (
    df[df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

print(gold_city_revenue)
```

```python
avg_order_by_cuisine = (
    df.groupby("cuisine")["total_amount"]
    .mean()
    .sort_values(ascending=False)
)

print(avg_order_by_cuisine)
```

```python
user_total_spend = (
    df.groupby("user_id")["total_amount"]
    .sum()
)

count_users = user_total_spend[user_total_spend > 1000].count()
print(count_users)
```

```python
df["rating_range"] = pd.cut(
    df["rating"],
    bins=[3.0, 3.5, 4.0, 4.5, 5.0],
    labels=["3.0–3.5", "3.6–4.0", "4.1–4.5", "4.6–5.0"]
)

rating_revenue = (
    df.groupby("rating_range")["total_amount"]
    .sum()
    .sort_values(ascending=False)
)

print(rating_revenue)
```

```python
gold_city_avg = (
    df[df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .mean()
    .sort_values(ascending=False)
)

print(gold_city_avg)
```

```python
restaurant_count = df.groupby("cuisine")["restaurant_id"].nunique()
revenue = df.groupby("cuisine")["total_amount"].sum()

summary = pd.concat([restaurant_count, revenue], axis=1)
summary.columns = ["restaurant_count", "total_revenue"]

print(summary.sort_values("restaurant_count"))
```

```python
gold_orders = df[df["membership"] == "Gold"].shape[0]
total_orders = df.shape[0]

percentage = round((gold_orders / total_orders) * 100)
print(percentage)
```
